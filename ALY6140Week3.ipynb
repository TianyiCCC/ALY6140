```python
from sqlalchemy import create_engine
engine = create_engine("sqlite:///./chinook.db")
engine.table_names()
```

```python
from sqlalchemy import inspect
inspector = inspect(engine)
```

```python
print(inspector.get_columns('tracks'))
print(inspector.get_table_names())
```

```python
import pandas as pd
df = pd.read_sql_query(
"""
select t.UnitPrice, t.GenreId
from
    tracks t

""", engine)
```

```python
print(df.head())
grouped = df.groupby('GenreId', as_index=False).mean()
grouped
```
